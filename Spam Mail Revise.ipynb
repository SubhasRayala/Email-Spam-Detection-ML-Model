```python
import pandas as pd
```

```python
df=pd.read_csv('spam.csv')
```

```python
df.head(5)
```

```python
df['Spam']=df.Category.apply(lambda x:1 if x=='spam' else 0)
```

```python
df
```

```python
df.drop('Category',axis='columns')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
from sklearn.model_selection import train_test_split
```

```python
inputs=df.Message
output=df.Spam
```

```python
X_train,X_test,y_train,y_test=train_test_split(inputs,output,test_size=0.2,random_state=15)
```

```python
X_train_count=cv.fit_transform(X_train)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
model1=MultinomialNB()
```

```python
model1.fit(X_train_count,y_train)
```

```python
X_test_count=cv.transform(X_test)
```

```python
model1.score(X_test_count,y_test)
```

# Using Pipeline

```python
from sklearn.pipeline import Pipeline
```

```python
pip=Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
pip.fit(X_train,y_train)
```

```python
pip.score(X_test,y_test)
```
